```python
Df = pd.read_csv("panda.csv")
Df.head()
```

```python
Df.info()
```

```python
Df.isnull()
```

```python
Df.isnull().sum()
```

```python
Df["Emp_id"].fillna("-",inplace = True)
Df["emp_name"].fillna("Unidentified",inplace = True)
Df["Salary"].fillna(Df["Salary"].mean(),inplace = True)
```

```python
Df.dropna(inplace = True)
Df
```

```python
Df.groupby("Dept").agg(
    {
        'Emp_id' : 'count',
        'Salary' : 'sum'
    }
)
```

```python
F_Dr = Df[Df["Salary"] >= 20000]
F_Dr
```

```python
F_Dr["Bonus"] = F_Dr["Salary"] // 10
F_Dr
```

```python
F_Dr["Emp_id"] = F_Dr["Emp_id"].astype(int)
F_Dr["Salary"] = F_Dr["Salary"].astype(int)
```
