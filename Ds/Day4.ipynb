```python
#matplotlib.pyplot
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'year': [2019,2020,2021,2022,2023,2024,2025],
    'sales':[3000,2400,6000,4000,7000,9000,10000]
}
df = pd.DataFrame(data)
df
```

```python
plt.plot(df['year'],df['sales'])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales in the Year 2019-25')
plt.show()
```

```python
import numpy as np
val_x = np.random.randint(5,40,size=30)
val_y = np.random.randint(50,100,size=30)
plt.scatter(val_x,val_y)
plt.title("Scatter plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
cinema = {
    'movies': ['Dheena','KGF2','Bahubali2','Gilli','Pokiri','Dragon'],
    'likes': [10000,12000,11000,10000,9500,13000]
}
cinema_df = pd.DataFrame(cinema)
cinema_df
```

```python
plt.bar(cinema_df['movies'],cinema_df['likes'],color="orange")
plt.title('Bar plot')
plt.xlabel('Movies')
plt.ylabel('Likes')
plt.show()
```

```python
val = np.random.randint(2,20,size=100)
plt.hist(val,color="green",edgecolor = "purple")

plt.title("histograph")
plt.show()
```

```python
import seaborn as sns
val1 = np.array([1,2,3,4,10,11,13,15,22,24,25,29,30,33,-100])
sns.boxplot(val1,color= "indigo")
plt.title("Boxplot")
plt.show()
```
